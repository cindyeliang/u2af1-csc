# Filter L4000 cpds list for cpds acting on specific pathways based on keyword

```python
#import pandas and np and wilcoxon ranksum calculator

import pandas as pd 
import numpy as np 
import scipy
from matplotlib.colors import ListedColormap
import scikit_posthocs
import statannot

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()

'''extract relevant columns from L4000'''

plate_df = pd.read_csv('L4000-Bioactive Compound Library-Beverley Rabbitts (KIT10006535).xlsx - Compound Information.tsv', 
                       sep='\t', lineterminator='\n')

working_df = plate_df[['Solvent', 'CAS', 'Name', 'Target', 'Pathways', 
                       'Receptor', 'Bioactivity']]

# print(working_df)
all_cpds_list = working_df.values.tolist() #convert df to list [ [solvnt, cas, name, target, pathway... ] , ...]

'''create a dataframe for cpds acting on user-specified pathway'''
pathway_phrase = str(input('enter pathway of interest to search for (case insensitive): '))

my_cpd_list = [] #initiialize empty cpds list for cpds of interest

'''if cpd pathway description has keyword entered, add that cpd info to the list'''
for row in all_cpds_list:
    if pathway_phrase.upper() in str(row[4]).upper():
        my_cpd_list.append(row)

'''turn selected cpd list to dataframe for ease of use'''
my_cpd_df = pd.DataFrame(my_cpd_list, columns = ['Solvent', 'CAS', 'Name', 'Target', 'Pathways', 'Receptor', 'Bioactivity'])
my_cpd_df
```
